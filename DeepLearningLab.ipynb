```python
import numpy as np
import matplotlib.pyplot as plt
```

Sigmoid function


```python
x = np.arange(-5,5,1)
y = 1 / (1 + np.exp(-x))
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(-5,5,1)
lambda_values = [1.5, 0.5, 1]

for lam in lambda_values:
    y = 1 / (1 + np.exp(-lam * x))
    plt.plot(x, y, label=f'Lambda = {lam}')



plt.show()
```

Relu function

```python
def relu(x):
  return np.maximum(0,x)

x = np.arange(-5,5,1)



plt.plot(x,relu(x))

plt.show()
```

Binary activation function

```python
def Binary_activation(x):
  return np.where(x>=0,1,0)

x = np.arange(-5,5,1)
plt.plot(x,Binary_activation(x))
plt.show()
```

tanh activation function

```python
def tanh_activation(x):
  return np.where(x>=10,1,0)

x = np.arange(-5,50,10)
plt.plot(x,tanh_activation(x))
plt.show()
```

soft max function

```python
import numpy as np
import matplotlib.pyplot as plt
def soft_max(x):
  return np.exp(x) / np.sum(np.exp(x))

x = np.arange(-5,5,1)
plt.plot(x,soft_max(x))
plt.show()
```

```python
import numpy as np
import pandas as pd

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Load data from Excel files
try:
  inputs = pd.read_csv('/content/Input_ABCD.csv').values
  labels = pd.read_csv('/content/Copy of Output_ABCD.csv').values # Flatten to 1D array
except FileNotFoundError:
  print("Error: Input_ABCD.xlsx or Label.csv not found. Please upload the files to your Colab environment.")
  exit()


inputs.shape
print(inputs)

# Get the first column of inputs and labels
inputs = inputs[:, 0]  # Select the first column of the input matrix
labels = labels[:, 0]  # Select the first column of the label matrix
# Initialize weights and bias
np.random.seed(0) # for reproducibility
weights = np.random.rand(len(inputs))
print("weights:" , weights)
bias = np.random.rand()

# Training parameters
learning_rate = 0.1
epochs = 1000

# Training loop
for _ in range(epochs):
    for i in range(len(inputs)):
        net = np.dot(inputs[i], weights) + bias
        output = sigmoid(net)
        error = labels[i] - output

        if (labels[i] != (output >= 0.5)).any():
          weights = weights + learning_rate * error * output * (1 - output) * inputs[i]
          bias = bias + learning_rate * error * output * (1 - output)

# Testing
print("Testing:")
for i in range(len(inputs)):
    net = np.dot(inputs[i], weights) + bias
    output = sigmoid(net)
    print(f"Input: {inputs[i]}, Output: {output:.4f}, Prediction: {1 if output >= 0.5 else 0}, Actual: {labels[i]}")

print(f"\nTrained weights: {weights}")
print(f"Trained bias: {bias}")
```
